```python
GRID_H,  GRID_W  = 13 , 13
BATCH_SIZE       = 16
```

```python
import tensorflow as tf
sess = tf.InteractiveSession()
```

```python



sess.close()
sess = tf.InteractiveSession()



cell_x = tf.to_float(tf.reshape(tf.tile(tf.range(GRID_W), [GRID_H]), (1, GRID_H, GRID_W, 1, 1)))


print(cell_x.shape)


print("")

a=tf.Print(cell_x,[cell_x],message="cell_x:")
a.eval()




#sess.close()

#print()
```

```python
sess.close()
sess = tf.InteractiveSession()


cell_y = tf.transpose(cell_x, (0,2,1,3,4))
print(cell_y.shape)

b=tf.Print(cell_y,[cell_y],message="cell_y:")
b.eval()
```

```python
sess.close()
sess = tf.InteractiveSession()

cell_grid = tf.tile(tf.concat([cell_x,cell_y], -1), [BATCH_SIZE, 1, 1, 5, 1])

print(cell_grid.shape)

c=tf.Print(cell_grid,[cell_grid],message="cell_grid:")
c.eval()
```
